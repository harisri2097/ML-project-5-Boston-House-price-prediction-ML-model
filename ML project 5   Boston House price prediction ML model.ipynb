```python
## Independent and Dependent features         #Splitting the data and Target

y = df['Price']                                                                    
x = df.drop(['Price'], axis=1)       
```

```python
from sklearn.model_selection import train_test_split               ### TRAIN TEST SPLIT           
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=3)                
```

```python
from sklearn.preprocessing import StandardScaler                                  
scaler=StandardScaler()                                                                      
```

```python
xtrain=scaler.fit_transform(xtrain)                          
xtest=scaler.transform(xtest)                     
```

```python
### Model Training
#1. LinearRegression
from sklearn.linear_model import LinearRegression
m1=LinearRegression()

m1.fit(xtrain,ytrain)  
```

```python
## print the coefficients and the intercept
print("COEF:",      m1.coef_)
print("INTERCEPT:", m1.intercept_)
print("WHICH PARAMETERS",m1.get_params())         ## on which parameters the model has been trained
```

```python
### Prediction With Test Data
reg_pred=m1.predict(xtest)
```

```python
Y_TEST=list(ytest)                

plt.plot(Y_TEST, color='blue', label = 'Actual Value')     
plt.plot(reg_pred, color='green', label='Predicted Value')            

plt.title('Actual Price vs Predicted Price')         
plt.xlabel('Number of values')
plt.ylabel('GLD Price')
plt.legend()
plt.show()
```

```python
###ASSUMPTION
## plot a scatter plot for the prediction
plt.scatter(ytest,reg_pred)
```

```python
## Residuals
residuals = ytest-reg_pred
print(residuals)
```

```python
## Plot this residuals 
sns.displot(residuals,kind="kde")
```

```python
## Scatter plot with respect to prediction and residuals
## uniform distribution
plt.scatter(reg_pred,residuals)
```

```python
# mean_squared_error, mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(ytest,reg_pred))
print(mean_squared_error(ytest,reg_pred))
print(np.sqrt(mean_squared_error(ytest,reg_pred)))
```

```python
## R square and adjusted R square
from sklearn.metrics import r2_score   
score=r2_score(ytest,reg_pred)
print(score)
```

```python
#display adjusted R-squared
1 - (1-score)*(len(ytest)-1)/(len(ytest)-xtest.shape[1]-1)
```

```python
#2. XGBoost Regressor
from xgboost import XGBRegressor
m2 = XGBRegressor()                                        # loading the model

m2.fit(xtrain, ytrain)                                   # training the model with X_train
```

```python
train_data_pred = m2.predict(xtrain)      ##Evaluation  #Prediction on training data  # accuracy for prediction on training data 
```

```python
# R squared error
score_1 = r2_score(ytrain, train_data_pred)
# Mean Absolute Error
score_2 = mean_absolute_error(ytrain, train_data_pred)              
print("R squared error : ",      score_1)
print('Mean Absolute Error :',   score_2)
```

```python
#Visualizing the actual Prices and predicted prices

plt.scatter(ytrain, train_data_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Preicted Price")
plt.show()
```
